## Task 1: Find Unique Elements from an Array
**Goal:** Create a function that takes a list and returns a new list containing only its unique elements.

```python
def get_unique_elements(input_list):
    """
    Takes a list and returns a new list containing only unique elements.
    Using a loop preserves the original order of first appearance.
    """
    unique_list = []
    for item in input_list:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list

# Test the function with the example input
example_input = [1, 2, 3, 2, 4, 1, 5]
output = get_unique_elements(example_input)

print("Input List: ", example_input)
print("Unique Elements: ", output)
```

## Task 2: Matrix Operations
**Goal:** Perform basic 2D array manipulations including matrix transposition and calculating diagonal sums.

```python
def transpose_matrix(matrix):
    """
    Returns the transpose of a given matrix (rows become columns).
    """
    rows = len(matrix)
    cols = len(matrix[0])
    
    # Initialize an empty matrix of dimensions (cols x rows)
    transposed = [[0 for _ in range(rows)] for _ in range(cols)]
    
    for i in range(rows):
        for j in range(cols):
            transposed[j][i] = matrix[i][j]
            
    return transposed

def calculate_diagonal_sums(matrix):
    """
    Calculates primary and secondary diagonal sums for a square matrix.
    """
    n = len(matrix)
    primary_sum = 0
    secondary_sum = 0
    
    for i in range(n):
        primary_sum += matrix[i][i]          # Elements like (0,0), (1,1), (2,2)
        secondary_sum += matrix[i][n - 1 - i] # Elements like (0,2), (1,1), (2,0)
        
    return primary_sum, secondary_sum

# ---- Testing Part A: Matrix Transpose ----
matrix_a = [
    [1, 2, 3],
    [4, 5, 6]
]
print("--- Part A: Transpose ---")
print("Original Matrix:")
for row in matrix_a: print(row)

transposed_result = transpose_matrix(matrix_a)
print("\nTransposed Matrix:")
for row in transposed_result: print(row)


# ---- Testing Part B: Diagonal Sums ----
matrix_b = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print("\n--- Part B: Diagonal Sums ---")
p_sum, s_sum = calculate_diagonal_sums(matrix_b)
print(f"Primary Diagonal Sum = {p_sum}")
print(f"Secondary Diagonal Sum = {s_sum}")
```

## Task 3: String Tokenization & Text Processing
**Goal:** Fetch external text data, clean it by removing punctuation, tokenize it into individual words, and analyze token frequency.

```python
import requests
import string

# 1. Fetching the dataset
url = "https://raw.githubusercontent.com/spyguessgame-boop/own_dataset/refs/heads/main/data.txt"
response = requests.get(url)
response.raise_for_status()

# Limit to the first 1000 characters as per instructions
text_data = response.text[:1000]
print("--- Raw Sample Text ---")
print(text_data[:300], "...\n") # Print a preview snippet

def clean_and_tokenize(text):
    """
    Removes punctuation, converts text to lowercase, splits into tokens, 
    and returns a list of words alongside basic analytics.
    """
    # Bonus Challenge: Remove punctuation using string.punctuation translation map
    # This replaces characters like periods, commas, and quotes with empty spaces
    clean_text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Standardize to lowercase so 'Word' and 'word' match
    clean_text = clean_text.lower()
    
    # Tokenize by splitting on whitespace
    tokens = clean_text.split()
    
    # Bonus Challenge: Total tokens count
    total_tokens = len(tokens)
    
    # Bonus Challenge: Find the most frequent token manually
    token_counts = {}
    for token in tokens:
        token_counts[token] = token_counts.get(token, 0) + 1
        
    most_frequent_token = None
    max_count = 0
    for token, count in token_counts.items():
        if count > max_count:
            max_count = count
            most_frequent_token = token
            
    return tokens, total_tokens, most_frequent_token, max_count

# Run the tokenization pipeline
tokens_list, count, frequent_word, frequency = clean_and_tokenize(text_data)

print("--- Tokenization Results ---")
print(f"First 10 generated tokens: {tokens_list[:10]}")
print(f"Total number of tokens: {count}")
print(f"Most frequent token: '{frequent_word}' (appears {frequency} times)")
```
