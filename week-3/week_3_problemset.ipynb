## Problem 1: Array Statistics
**Goal:** Calculate the total sum, mean average, and maximum value of a 1D NumPy array without built-in loops.

```python
import numpy as np

# Input array
arr = np.array([1, 2, 3, 4, 5])

# Calculate statistics using NumPy methods
arr_sum = np.sum(arr)
arr_mean = np.mean(arr)
arr_max = np.max(arr)

print(f"Sum = {arr_sum}")
print(f"Mean = {arr_mean}")
print(f"Max = {arr_max}")
```

## Problem 2: Even Number Filter
**Goal:** Filter out odd elements and return only even integers using NumPy boolean masking.

```python
arr_even = np.array([1, 2, 3, 4, 5, 6])

# Apply boolean indexing to filter even numbers
even_filtered = arr_even[arr_even % 2 == 0]

print(even_filtered.tolist())
```

## Problem 3: Row With Maximum Sum
**Goal:** Compute row-wise sums on a 2D array and pinpoint the precise index of the highest summing row.

```python
matrix_3 = np.array([
    [1, 2, 3],
    [5, 5, 5],
    [2, 2, 2]
])

# Sum across rows (axis=1) and find the index of the maximum value
row_sums = np.sum(matrix_3, axis=1)
max_row_index = np.argmax(row_sums)

print(f"Row sums = {row_sums.tolist()}")
print(f"Index of row with max sum: {max_row_index}")
```

## Problem 4: Student Performance Analysis
**Goal:** Find the student indices whose average score exceeds 75 across multiple subjects.

```python
marks = np.array([
    [80, 90, 70],
    [50, 60, 70],
    [90, 95, 85]
])

# Calculate average marks per student (axis=1 means row-wise mean)
student_averages = np.mean(marks, axis=1)

# Find indices where average marks > 75
passing_students = np.where(student_averages > 75)[0]

print(f"Student Averages: {student_averages.tolist()}")
print(f"Passing Student Indices: {passing_students.tolist()}")
```

## Problem 5: Network Traffic Anomaly Detection (Very Hard Version)
**Goal:** Identify specific server indices where localized traffic spikes past twice the entire network global mean average.

```python
traffic = np.array([
    [100, 120, 110],
    [90, 95, 100],
    [500, 550, 600]
])

# 1. Calculate the overall average across ALL servers and data points
overall_average = np.mean(traffic)

# 2. Calculate the mean traffic for each individual server (row-wise mean)
server_averages = np.mean(traffic, axis=1)

# 3. Use np.where to check which servers exceed 2 * overall_average
anomalous_servers = np.where(server_averages > 2 * overall_average)[0]

print(f"Overall Network Average: {overall_average:.2f}")
print(f"Anomalous Server Indices: {anomalous_servers.tolist()}")
```

## Problem 6: Fraud Detection Feature Generator (Very Hard Version)
**Goal:** Normalize disparate datasets (Min-Max normalization), compute weighted composite risk scores, and isolate indices of the top $K$ highest-risk users.

```python
transactions = np.array([
    [1000, 50, 0],
    [5000, 20, 2],
    [3000, 80, 5]
])
k = 2

# Min-Max Normalization: (X - min) / (max - min) column-by-column
# np.min and np.max with axis=0 operates down the column features
col_min = np.min(transactions, axis=0)
col_max = np.max(transactions, axis=0)

# Vectorized scaling: handling potential zero division just in case
normalized_tx = (transactions - col_min) / (col_max - col_min)

# Weights vector as structured in the risk score formula
weights = np.array([0.6, 0.3, 0.1])

# Calculate risk score for each user via matrix-vector dot product multiplication
risk_scores = np.dot(normalized_tx, weights)

# Sort based on risk index. np.argsort sorts ascending; [::-1] reverses it to descending
top_k_indices = np.argsort(risk_scores)[::-1][:k]

print("--- Fraud System Diagnostics ---")
for i, score in enumerate(risk_scores):
    print(f"User {i} -> Normalized Metrics: {normalized_tx[i].round(2)} | Final Risk Score: {score:.2f}")

print(f"\nTop {k} Most Risky User Indices: {top_k_indices.tolist()}")
```
